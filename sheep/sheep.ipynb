```python
from datetime import timedelta, datetime
from dateutil.relativedelta import relativedelta

REPRODUCTION_EFFICIENCY = 0.85
FIRST_AVG_OFFSPRING = 1
AVG_OFFSPRING = 1.5
RAM_REV_USD = 335

class SheepGeneration():

    def __init__(self, generation, n, dob):
        self.generation = generation
        self.n_sheep = n
        self.dob = dob

        self.age = 0 # in months
        

    def __str__(self):
        return f"Generation: {self.generation}, Age: {self.age / 12:.2f} y, Sheep: {self.n_sheep}, DOB: {str(self.dob)[:10]}"

    def make_offspring(self):
        # returns the number of ewe lambs expected to be born (which is the same as the number of rams)
        ewes = 0
        if self.age <= 13:
            ewes = FIRST_AVG_OFFSPRING * self.num_ewes * REPRODUCTION_EFFICIENCY
        else:
            ewes = AVG_OFFSPRING * self.num_ewes * REPRODUCTION_EFFICIENCY

        ewe_lambs = 0.5 * ewes

        return ewe_lambs


    def increase_age(self):
        self.age += 1


    def can_reproduce(self):
        if self.age >= 84:
            return False

        if self.age < 12: 
            return False
        elif self.age == 12:
            return True
        elif (self.age - 12) % 7 == 0:
            # every 7 months after 12 months
            return True


    def merge(self, other):
        assert self.dob == other.dob

        self.num_ewes += other.num_ewes
        self.num_rams += other.num_rams

        return self



        

    
```

```python
# create first gen of sheep
today = datetime(2022, 12, 1)

first_gen = SheepGeneration("YGB", 22, today)
generations = [first_gen]

gen_counter = 1

generation_dict = dict()

while today < datetime(2030, 12, 1):
    today += relativedelta(months=1)
    
    for generation in generations:
        generation.increase_age()

        if generation.can_reproduce():
            new_gen = SheepGeneration(f"YGB-{gen_counter}", generation.make_offspring(), today)

            if today in generation_dict:
                generation_dict[today].merge(new_gen)
            else:
                generation_dict[today] = new_gen
            
            generations.append(new_gen)

            gen_counter += 1

#second_gen = SheepGeneration("YGB-2", first_gen.make_offspring())
```
